## Меры центральной тенденции

### Мода
Мода - это значение измеряемового признака, которое встречается максимальное число раз. Мод может быть несколько, если несколько значений имеют одинаковое максимальное число вхождений в выборку.



```python
data = np.random.randint(150,200,100)

df = pd.DataFrame(data, columns = ["Рост"])
df
```

Пускай эта рандомная выборка - это рост опрошенных людей, найдём моду, для этого:

```python
mode = df["Рост"].mode()[0]
```

```python
plt.figure(figsize=(20, 12))
sns.countplot(df, x = "Рост")
plt.xlabel("Рост")
plt.ylabel("Количество")

plt.axhline(y=df['Рост'].value_counts().max(), color='red', linestyle='--')
plt.text(x=-0.5, y=df['Рост'].value_counts().max() + 0.5, s=f'Мода: {mode}', color='red')
```

### Медиана
Медиана - это значение признака, для которого в отсортированном по возрастанию наборе выполняется следующее:
Половина значений находиться слева от медианы и половина справа.

```python
median = df["Рост"].median()
```

```python
counts, bins= np.histogram(data,bins = 20)

plt.figure(figsize=(16, 10))
plt.stairs(counts, bins)

plt.axvline(x=median, color='red', linestyle='--')
```

### Среднее значение
Среднее значение - это сумма всех значений деленное на их количество

```python
mean = df["Рост"].mean()
```

```python
plt.figure(figsize=(16, 10))
plt.stairs(counts, bins)

plt.axvline(x=mean, color='red', linestyle='--')
```
