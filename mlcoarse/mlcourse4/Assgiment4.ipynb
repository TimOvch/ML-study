```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Lasso, LassoCV, LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import (GridSearchCV, cross_val_score,
                                     train_test_split)
from sklearn.preprocessing import StandardScaler
```

```python
# for Jupyter-book, we copy data from GitHub, locally, to save Internet traffic,
# you can specify the data/ folder from the root of your cloned
# https://github.com/Yorko/mlcourse.ai repo, to save Internet traffic
DATA_PATH = "https://raw.githubusercontent.com/Yorko/mlcourse.ai/main/data/"

data = pd.read_csv(DATA_PATH + "winequality-white.csv", sep=";")
```

```python
data.head()
```

```python
data.info()
```

```python
y = data['quality'] 
X = data.drop('quality', axis =  1)

X_train, X_holdout, y_train, y_holdout = train_test_split(X,y, test_size=0.3,random_state = 17)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train.values.tolist())
X_holdout_scaled = scaler.transform(X_holdout.values.tolist())
```

```python
# (read-only in a JupyterBook, pls run jupyter-notebook to edit)
linreg = LinearRegression()
linreg.fit(X_train_scaled, y_train)
```

```python
# (read-only in a JupyterBook, pls run jupyter-notebook to edit)

MSE_test = ((y_holdout - linreg.predict(X_holdout_scaled))**2).mean()
MSE_train = ((y_train - linreg.predict(X_train_scaled))**2).mean()
print("Mean squared error (train): " + str(MSE_train))
print("Mean squared error (test): " + str(MSE_test))
```

```python
# (read-only in a JupyterBook, pls run jupyter-notebook to edit)
linreg_coef = pd.DataFrame(
    {"coef": linreg.coef_, "coef_abs": np.abs(linreg.coef_)},
    index=data.columns.drop("quality"),
)
linreg_coef.sort_values(by="coef_abs", ascending=False)
```

```python
# (read-only in a JupyterBook, pls run jupyter-notebook to edit)
lasso1 = Lasso(alpha=0.01)
lasso1.fit(X_train_scaled, y_train)
```

```python
# (read-only in a JupyterBook, pls run jupyter-notebook to edit)
lasso1_coef = pd.DataFrame({'coef' : lasso1.coef_, 'coef_abs' : abs(lasso1.coef_)}, index = data.drop('quality', axis = 1).columns)
lasso1_coef.sort_values(by = 'coef_abs', ascending = False)
```

```python
# (read-only in a JupyterBook, pls run jupyter-notebook to edit)
%time

alphas = np.logspace(-6, 2, 200)
lasso_cv = LassoCV(alphas=alphas,cv=5, n_jobs=-1,random_state=17)
lasso_cv.fit(X_train_scaled, y_train)
```

```python
lasso_cv.alpha_
```

```python
lasso_cv_coef = pd.DataFrame({'coef' : lasso_cv.coef_, 'coef_abs' : abs(lasso_cv.coef_)}, index = data.drop('quality', axis = 1).columns)
lasso_cv_coef.sort_values(by = 'coef_abs', ascending = False)
```

```python
# (read-only in a JupyterBook, pls run jupyter-notebook to edit)
print("Mean squared error (train): %.3f" % mean_squared_error(lasso_cv.predict(X_train_scaled), y_train))
print("Mean squared error (train): %.3f" % mean_squared_error(lasso_cv.predict(X_holdout_scaled), y_holdout))
```

```python
# (read-only in a JupyterBook, pls run jupyter-notebook to edit)
forest = RandomForestRegressor(random_state=17)
forest.fit(X_train, y_train)
```

```python
# (read-only in a JupyterBook, pls run jupyter-notebook to edit)
print("Mean squared error (train): %.3f" % mean_squared_error(forest.predict(X_train), y_train))
print("Mean squared error (train): %.3f" % mean_squared_error(forest.predict(X_holdout), y_holdout))
```

```python
%%time
forest_params = {'max_depth': list(range(10, 25)),
                 'min_samples_leaf': list(range(1, 8)),
                 'max_features': list(range(6,12))}

locally_best_forest = GridSearchCV(estimator=forest, n_jobs=-1, cv=5, param_grid =forest_params)
locally_best_forest.fit(X_train, y_train)
```

```python
locally_best_forest.best_params_, locally_best_forest.best_score_
```

```python
# (read-only in a JupyterBook, pls run jupyter-notebook to edit)
print(
    "Mean squared error (cv): %.3f"
    % np.mean(
        np.abs(
            cross_val_score(
                locally_best_forest.best_estimator_,
                X_train,
                y_train,
                scoring="neg_mean_squared_error",
            )
        )
    )
)
print(
    "Mean squared error (test): %.3f"
    % mean_squared_error(y_holdout, locally_best_forest.predict(X_holdout))
)
```

```python
# (read-only in a JupyterBook, pls run jupyter-notebook to edit)
rf_importance = pd.DataFrame({'imp' : locally_best_forest.best_estimator_.feature_importances_, 
                              'imp_abs' : np.abs(locally_best_forest.best_estimator_.feature_importances_)},
                            index = data.drop('quality', axis = 1 ).columns)
rf_importance.sort_values(by = 'imp_abs', ascending = False)
```
